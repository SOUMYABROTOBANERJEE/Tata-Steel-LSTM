```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data16= pd.read_excel('Book1.xlsx')
print(data16)
print('the shape of our features is:', data16.shape)
data16.describe()
data16=pd.get_dummies(data16)
labels=np.array(data16['Status_Overdue'])
data16=data16.drop('Status_Overdue', axis=1)
data16=data16.drop('Status_Not OD',axis=1)
data16=data16.drop('Arr (Clearing - Net Due Date)',axis=1)
data16.dropna(inplace=True)
data16_list=list(data16.columns)
#data16=np.array(data16)
from sklearn.model_selection import train_test_split
train_data16, test_data16, train_labels, test_labels= train_test_split(data16, labels, test_size=0.25, random_state=0)
print('training data16 shape:', train_data16.shape)
print('training labels shape:', train_labels.shape)
print('testing data16 shape:', test_data16.shape)
print('testing labels shape:', test_labels.shape)
#base_preds=test_data16[:, data16_list.index('  Local Crcy Amt')]
#base_errors=abs(base_preds- test_labels)
#print('average baseline error:', round(np.mean(base_errors),2))
from sklearn.tree import DecisionTreeClassifier
rf1=DecisionTreeClassifier(criterion="entropy", max_depth=4)
rf1.fit(train_data16, train_labels)
predictions1=rf1.predict(test_data16)
from sklearn.metrics import confusion_matrix, classification_report, precision_score
print(classification_report(test_labels, predictions1))
errors=abs(predictions1 - test_labels)
print('mean absolute error:', round(np.mean(errors), 2), 'degrees.')
#sums=sum(errors)
#print(sums)
#sum1=sum(test_labels)
#print(sum1)
mape=100*(errors)
accuracy=100-np.mean(mape)
print('accuracy:', round(accuracy, 2), '%')
```
